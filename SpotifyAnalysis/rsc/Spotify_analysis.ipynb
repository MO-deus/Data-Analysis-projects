```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mptlib
import seaborn as sb
```

```python
# 
df_tracks = pd.read_csv("D:/dev/Data_analysis/SpotifyAnalysis/rsc/tracks.csv")
```

```python
#checkings for null values
pd.isnull(df_tracks).sum()
```

```python
df_tracks.describe().transpose()
```

```python
# finding 10 most popular songs 
most_pop = df_tracks.query('popularity>90', inplace=False).sort_values('popularity',ascending=False)
most_pop[:10]

#  
```

```python
# setting the index as date_time 
df_tracks.set_index('release_date', inplace=True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
# checking a specific entry in ith index
df_tracks[["artists"]].iloc[18]
```

```python
# changing the duration from milisecond to seconds
df_tracks["duration_s"] = df_tracks["duration_ms"].apply(lambda x : (x/1000))
df_tracks.drop("duration_ms", inplace=False, axis=1)    
df_tracks.duration_s.head()
```

```python
corr_df = df_tracks.drop(["key","mode","explicit"], axis = 1).corr(method = "pearson")
mptlib.figure(figsize=(14, 6))
heatmap = sb.heatmap(corr_df, annot=True, fmt = ".1g", vmin=-1,vmax=1, center=0, cmap="inferno", linecolor="Black")
heatmap.set_title("Correlation heatmap between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

```python
sample_df = df_tracks.sample(int(0.004 * len(df_tracks)))
print(len(sample_df))
```

```python
mptlib.figure(figsize= (10,6))
sb.regplot(data=sample_df, y="loudness", x="energy", color="c").set(title = "Loudness vs energy Correlation")
```

```python
mptlib.figure(figsize= (10,6))
sb.regplot(data=sample_df, y="popularity", x="acousticness", color="y").set(title = "popularity vs acousticness Correlation")
```

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
sb.displot(data=years, discrete = True, aspect=2, height=5, kind='hist').set(title = "Number of songs per year")
```

```python
print("info on sample_df : 4% of actual data")
sample_df.info()

print("info on sample_df : 4% of actual data")
df_tracks.info()

```

```python
total_dr = df_tracks.duration_ms
fig_dims = (18,7)
fig, ax = mptlib.subplots(figsize = fig_dims)
fig = sb.barplot(x=years, y = total_dr, ax = ax ,errwidth=False).set(title = "Barplot against years and total duration")
mptlib.xticks(rotation = 90)
```

```python
total_dr = df_tracks.duration_ms
sb.set_style(style="whitegrid")
fig_dims = (10,5)
fig, ax = mptlib.subplots(figsize = fig_dims)
fig = sb.lineplot(x=years ,y=total_dr, ax=ax).set(title = "Lineplot against years and duration")
mptlib.xticks(rotation = 60)
```
